```python
import matplotlib.pyplot as plt
import numpy as np

# Define the pathologies and their corresponding AUC values from the CheXNeXt model and radiologists
pathologies = [
    'Atelectasis', 'Cardiomegaly', 'Consolidation', 'Edema', 'Effusion',
    'Emphysema', 'Fibrosis', 'Hernia', 'Infiltration', 'Mass',
    'Nodule', 'Pleural thickening', 'Pneumonia', 'Pneumothorax'
]

auc_values_chexnext = [
    0.862, 0.831, 0.893, 0.924, 0.901,
    0.704, 0.806, 0.851, 0.721, 0.909,
    0.894, 0.798, 0.851, 0.944
]

auc_values_radiologists = [
    0.808, 0.888, 0.841, 0.910, 0.900,
    0.911, 0.897, 0.985, 0.734, 0.886,
    0.899, 0.779, 0.823, 0.940
]

# Plotting the AUC values for the CheXNeXt model and Radiologists
fig, ax = plt.subplots(figsize=(12, 8))

index = np.arange(len(pathologies))
bar_width = 0.35

bar1 = plt.barh(index - bar_width/2, auc_values_chexnext, bar_width, label='CheXNeXt Model', color='skyblue')
bar2 = plt.barh(index + bar_width/2, auc_values_radiologists, bar_width, label='Radiologists', color='orange')

plt.xlabel('AUC Value')
plt.title('AUC values for the CheXNeXt model and radiologists on the dataset')
plt.yticks(index, pathologies)
plt.xlim(0, 1)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adding AUC values on the bars
for i, v in enumerate(auc_values_chexnext):
    plt.text(v + 0.01, i - bar_width/2, f'{v:.3f}', va='center', fontsize=9)
for i, v in enumerate(auc_values_radiologists):
    plt.text(v + 0.01, i + bar_width/2, f'{v:.3f}', va='center', fontsize=9)

# Place the legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```
